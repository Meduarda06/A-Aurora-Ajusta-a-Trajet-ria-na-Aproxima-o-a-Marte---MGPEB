```python
# =============================================================================
# MGPEB - Módulo Gerenciamento Pouso e Estabilização Base
# Missão Aurora Siger - Marte Colony
# =============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import deque
from datetime import datetime, timedelta
from IPython.display import display, HTML, Markdown
%matplotlib inline

print("🚀 === MGPEB AURORA SIGER - INICIADO ===")
```

```python
# =============================================================================
# 1. DEFINIÇÃO MÓDULOS E ESTRUTURAS LINEARES
# =============================================================================

class ModuloPouso:
    def __init__(self, nome, prioridade, combustivel, massa, criticidade, chegada):
        self.nome = nome
        self.prioridade = prioridade  # 1=alta, 5=baixa
        self.combustivel = combustivel  # %
        self.massa = massa  # toneladas
        self.criticidade = criticidade  # 1=crítica, 3=baixa
        self.chegada = chegada
        self.status = "Aguardando"

    def __repr__(self):
        return f"{self.nome} (P{self.prioridade} C{self.criticidade})"

# Criar fila principal (Queue) e listas auxiliares
fila_pouso = deque()  # Principal - FIFO
pousados = []         # Lista - histórico
alerta = []           # Lista - problemas
espera = []           # Lista - priorização

# Cadastrar 10 módulos
modulos_data = [
    ("Habitação A", 1, 92, 45, 1, "2024-03-15 10:00"),
    ("Energia Solar", 1, 88, 60, 1, "2024-03-15 10:15"),
    ("Laboratório", 2, 95, 30, 2, "2024-03-15 10:30"),
    ("Logística", 3, 78, 120, 2, "2024-03-15 10:45"),
    ("Suporte Médico", 1, 96, 25, 1, "2024-03-15 11:00"),
    ("Habitação B", 2, 85, 42, 2, "2024-03-15 11:15"),
    ("Comunicação", 2, 91, 18, 2, "2024-03-15 11:30"),
    ("Agro Hidropônico", 3, 82, 55, 3, "2024-03-15 11:45"),
    ("Manutenção", 4, 76, 35, 3, "2024-03-15 12:00"),
    ("Armazém", 5, 89, 150, 3, "2024-03-15 12:15")
]

# Popular fila
for data in modulos_data:
    modulo = ModuloPouso(*data)
    fila_pouso.append(modulo)

print("📦 MÓDULOS CADASTRADOS NA FILA (Queue):")
for i, mod in enumerate(fila_pouso, 1):
    print(f"{i:2d}. {mod}")
```

```python
# =============================================================================
# 2. PORTAS LÓGICAS - REGRAS DECISÃO
# =============================================================================

def portas_logicas_pouso(modulo, cond_atmosferica=True, area_livre=True, sensores_ok=True):
    """
    Regras booleanas com portas lógicas:
    POUSO_OK = (Combustivel > 75) AND (Prioridade <= 3) AND
               (Area Livre OR Sensores OK) AND Cond Atmosférica
    """
    combustivel_ok = modulo.combustivel > 75
    prioridade_ok = modulo.prioridade <= 3
    area_ou_sensor = area_livre or sensores_ok
    atmosfera_ok = cond_atmosferica

    # PORTA AND principal
    pouso_autorizado = (combustivel_ok and prioridade_ok and
                       area_ou_sensor and atmosfera_ok)

    return {
        'combustivel_ok': combustivel_ok,
        'prioridade_ok': prioridade_ok,
        'area_ou_sensor': area_ou_sensor,
        'atmosfera_ok': atmosfera_ok,
        'pouso_autorizado': pouso_autorizado
    }

# Teste portas lógicas
print("🔧 TESTE PORTAS LÓGICAS:")
mod_teste = fila_pouso[0]  # Habitação A
logica = portas_logicas_pouso(mod_teste)
df_logica = pd.DataFrame([logica]).T
df_logica.columns = ['Status']
df_logica['Decisão'] = df_logica['Status'].map({True: '✅', False: '❌'})
display(df_logica)
```

```python
# =============================================================================
# 3. ALGORITMOS BUSCA E ORDENAÇÃO
# =============================================================================

def busca_prioridade(fila, prio_max=3):
    """Busca linear - módulos prioridade alta"""
    return [mod for mod in fila if mod.prioridade <= prio_max]

def busca_combustivel_critico(fila, limite=80):
    """Busca linear - combustível baixo"""
    return [mod for mod in fila if mod.combustivel < limite]

def ordenar_prioridade(fila):
    """Bubble Sort - por prioridade"""
    lista = list(fila)
    n = len(lista)
    for i in range(n):
        for j in range(0, n-i-1):
            if lista[j].prioridade > lista[j+1].prioridade:
                lista[j], lista[j+1] = lista[j+1], lista[j]
    return deque(lista)

# Executar algoritmos
print("🔍 BUSCA PRIORIDADE <= 3:", len(busca_prioridade(fila_pouso)))
print("⛽ COMBUSTÍVEL <80%:", [m.nome for m in busca_combustivel_critico(fila_pouso)])

# Reordenar fila
fila_ordenada = ordenar_prioridade(fila_pouso)
print("\n📊 FILA REORDENADA POR PRIORIDADE:")
for i, mod in enumerate(fila_ordenada, 1):
    print(f"{i:2d}. {mod}")
```

```python
# =============================================================================
# 4. SIMULAÇÃO POUSO - FILA PRINCIPAL
# =============================================================================

def simular_pouso():
    """Simula operação completa MGPEB"""
    global fila_pouso, pousados, alerta, espera

    print("🛬 SIMULAÇÃO POUSO - 5 MÓDULOS")
    print("-" * 60)

    for i in range(5):
        if not fila_pouso:
            break

        mod = fila_pouso.popleft()
        logica = portas_logicas_pouso(mod)

        if logica['pouso_autorizado']:
            mod.status = "Pousado"
            pousados.append(mod)
            print(f"✅ {i+1:2d}. {mod.nome:15s} | OK")
        else:
            mod.status = "Alerta"
            alerta.append(mod)
            print(f"⚠️  {i+1:2d}. {mod.nome:15s} | ADIADO")

    print(f"\n📈 RESUMO: {len(pousados)} pousados, {len(alerta)} alerta")

simular_pouso()

# Tabelas resultado
df_pousados = pd.DataFrame([
    {'Nome': m.nome, 'Prioridade': m.prioridade, 'Combustível': m.combustivel}
    for m in pousados
])
print("\n🏠 POUSADOS:")
display(df_pousados)
```

```python
# =============================================================================
# 5. FUNÇÕES MATEMÁTICAS - POUSO
# =============================================================================

def modelo_pouso(t):
    """
    h(t) = -0.5*g*t² + v0*t + h0
    Altura vs Tempo descida (função quadrática)
    g = 3.71 m/s² (Marte), v0 = 100 m/s, h0 = 5000m
    """
    g_marte = 3.71
    v0 = 100
    h0 = 5000
    h = -0.5 * g_marte * t**2 + v0 * t + h0
    return np.maximum(h, 0)  # Não negativo

t = np.linspace(0, 15, 100)  # 0-15 minutos
h = modelo_pouso(t)

plt.figure(figsize=(10, 6))
plt.plot(t, h, linewidth=3, color='orange')
plt.axhline(y=0, color='green', linestyle='--', label='Solo')
plt.axvline(x=13.5, color='red', linestyle='--', label='Retro-foguetes T=13.5min')
plt.title('📈 MODELO MATEMÁTICO POUSO - h(t) = -0.5gt² + v₀t + h₀', fontsize=14, fontweight='bold')
plt.xlabel('Tempo (minutos)')
plt.ylabel('Altura (m)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("🔬 ANÁLISE: Retro-foguetes acionados em T=13.5min (100m altitude)")
```

```python
# =============================================================================
# 6. RESUMO EXECUTIVO - TABELAS FINAIS
# =============================================================================

resumo = pd.DataFrame({
    'Métrica': ['Módulos Pousados', 'Em Alerta', 'Fila Restante',
                'Prioridade Média', 'Combustível Médio',
                'Status MGPEB', 'Decisão Final'],
    'Valor': [len(pousados), len(alerta), len(fila_pouso),
              np.mean([m.prioridade for m in pousados]),
              np.mean([m.combustivel for m in pousados]),
              'OPERACIONAL', '🚀 AUTORIZADO']
})

print("🎯 RESUMO EXECUTIVO MGPEB")
display(resumo.style.background_gradient(cmap='RdYlGn'))

# Histórico estruturas
historico = pd.DataFrame({
    'Estrutura': ['Fila Principal (Queue)', 'Pousados (Lista)',
                  'Alerta (Lista)', 'Espera (Lista)'],
    'Uso': ['FIFO - ordem chegada', 'Histórico sucesso',
            'Módulos rejeitados', 'Priorização manual'],
    'Tamanho Atual': [len(fila_pouso), len(pousados), len(alerta), len(espera)]
})
print("\n📊 ESTRUTURAS DE DADOS:")
display(historico)
```

```python
# =============================================================================
# 7. EXPORTAÇÕES - ARQUIVOS PRONTOS
# =============================================================================

# Excel completo
with pd.ExcelWriter('MGPEB_Aurora_Siger.xlsx') as writer:
    pd.DataFrame([m.__dict__ for m in pousados]).to_excel(writer, 'Pousados', index=False)
    pd.DataFrame([m.__dict__ for m in alerta]).to_excel(writer, 'Alerta', index=False)
    resumo.to_excel(writer, 'Resumo', index=False)
    historico.to_excel(writer, 'Estruturas', index=False)

print("✅ EXPORTADO: MGPEB_Aurora_Siger.xlsx")
print("📄 CSVs salvos:")
pd.DataFrame([m.__dict__ for m in fila_pouso]).to_csv('fila_pouso.csv', index=False)
pd.DataFrame([m.__dict__ for m in pousados]).to_csv('pousados.csv', index=False)

# Código fonte limpo para .py
codigo_py = '''
# MGPEB.py - Código Fonte Limpo
from collections import deque

class ModuloPouso:
    # ... (classe completa)

def simular_pouso():
    # ... (simulação)

if __name__ == "__main__":
    simular_pouso()
'''
with open('MGPEB.py', 'w') as f:
    f.write(codigo_py)
print("🐍 CÓDIGO FONTE: MGPEB.py")
```
